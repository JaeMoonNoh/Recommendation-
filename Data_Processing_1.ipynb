```python
def preprocessing_data(webtoon):
  webtoon.drop(['Unnamed: 0','id','author','day','platform'], axis = 1, inplace = True)
  webtoon['title_genre_story'] = webtoon['title'] + " " + webtoon['genre'] + " " + webtoon['story']
  webtoon.drop(['story'], axis = 1, inplace = True)
  return webtoon 

#데이터 전처리 방법
```

```python
webtoon = pd.read_csv('네이버_웹툰_2.csv')
webtoon.head()
```

```python
webtoon['story'] = webtoon['story'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
webtoon.head()
```

```python
preprocessing_data(webtoon)
webtoon.head()
```

```python
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

komoran = Komoran()
td = []
for sentence in webtoon['title_genre_story']:
  temp_X = komoran.nouns(sentence)
  temp_X = [word for word in temp_X if not word in stopwords]
  td.append(temp_X)

webtoon['komoran'] = td
```
